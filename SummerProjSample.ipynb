```python
# Importing Libraries

import pandas as pd
import numpy as np 
import random
import math
```

```python
# Reading files

peo = pd.read_csv("People.csv")
cr = pd.read_csv("Cars.csv")
hs = pd.read_csv("Houses.csv")
```

```python
# Approximating function

def rounder(n):
    l = len(str(n))
    n = math.ceil(n / (math.pow(10, l-1)))
    n = n * math.pow(10, l-1)
    return int(n)
```

```python
# Setting the "مخارج ماهیانه هر نفر در خانواده" column

cost = []

for i in range(20):
    cost.append(rounder(random.randint(10000000, 50000000)))

peo['مخارج ماهیانه هر نفر در خانواده'] = cost

peo
```

```python
# Setting the "تراز مالی ماهیانه"

peo['تراز مالی ماهیانه'] = peo['درآمد ماهیانه'] - peo['مخارج ماهیانه هر نفر در خانواده'] * peo['تعداد اعضا خانواده']

peo
```

```python
# Setting the "قیمت خودرو"

cars_price = []
for cars in peo['خودرو']:
    cars_price.append(int(cr['قیمت خودرو'].loc[cr['نام خودرو']==cars]))
peo['قیمت خودرو'] = cars_price

peo
```

```python
# Setting the "قیمت خانه"

houses_price = []
houses = peo['منطقه مسکونی']
count = peo['متراژ خانه']
for i in range(20):
    houses_price.append(int(hs['قیمت هر متر مربع'].loc[hs['منطقه مسکونی']==houses[i]]) * count[i])
    
peo['قیمت خانه'] = houses_price

peo 
```

```python
# Setting the "سرمایه"

peo['سرمایه'] = peo['قیمت خانه'] + peo['قیمت خودرو'] + (peo['تراز مالی ماهیانه'] * 12 * 5)

peo
```

```python
# Calculating

a = peo['سرمایه'].mean()
b = peo['تراز مالی ماهیانه'].mean() * 12 * 5
namirayi = a + b        #69318000000.
first, second, third = (namirayi*1/2), (namirayi), (namirayi*3/2)

print(first, second, third)
```

```python
# Grouping the people and setting the "چهارک"

groups = []

for s in peo['سرمایه']:
    if s<first:
        groups.append(1)
    elif s<second:
        groups.append(2)
    elif s<third:
        groups.append(3)
    else:
        groups.append(4)

peo['چهارک'] = groups

peo
```

```python
# Exporting the data

peo.to_csv("FinalData.csv") 
```
